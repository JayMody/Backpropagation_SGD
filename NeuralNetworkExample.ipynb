# Feed Forward
The feed forward process for our network would look like this,

\begin{align}
\large
X \xrightarrow[W^1X + B^1]{\text{linear}} h_i \xrightarrow[f(h_i)]{\text{activation}} h_o \xrightarrow[W^2h_o + B^2]{\text{linear}} y_i \xrightarrow[f(y_i)]{\text{activation}} \hat{y}
\end{align}

where:
- $X$ is our inputs
- $W^1$ is our weight matrix connecting the input layer to the hidden layer
- $B^1$ is the bais vector for the input layer to the hidden layer
- $W^2$ is our weight matrix connecting the hidden layer to the output layer
- $B^2$ is the bais vector for the hidden layer to the output layer
- $f(x)$ is an activation function

The statement below each arrow represents the function being applied to the value to the left of it, where the output of that function is the value to the right of it. So for example:

\begin{align}
\large
X \xrightarrow[W_1X + B_1]{\text{linear}} h_i
\end{align}

means that:

\begin{align}
\large
h_i = W^1X + B^1
\end{align}


### Activation (Sigmoid)
The activation function we will be using is the sigmoid (logistic) function:

\begin{align}
\large
\sigma (x) = \frac{1}{1+e^{-x}}
\end{align}

# Error 
\begin{align}
\large
network \xrightarrow[]{} \hat{y} \xrightarrow[E(\hat{y}, y)]{} error
\end{align}

The predictions and labels are put through a loss funciton. The result is the error of the inputs.

### Error Function (Mean Squared Error)
The error function we will be using is mean squared error (MSE):

\begin{align}
\large
E(y, \hat{y}) = (y - \hat{y})^2
\end{align}

For some context, in the gradient descent step, the individual squared errors of the prediction-label pairs is what  is being backpropagated, not the MEAN squared error. So if you had 4 output nodes, you would have 4 seperate errors for each output node, not the same mean error. The mean is only used when measuring the error of the predictions (not involved in the backpropagation step). This makes sense, since if you treated every output node with the mean error, you are losing information (if one node has low error, but the other had high error, if you took the mean, they would be treated the same)

Here's the full loss function:

\begin{align}
\large
E(y, \hat{y}) = \frac{1}{N} \sum_{i=1}^N (y_i - \hat{y_i})^2
\end{align}

And for simplicities sake, we can also add a 1/2 constant to make derivation later on cleaner:

\begin{align}
\large
E(y, \hat{y}) = \frac{1}{2N} \sum_{i=1}^N (y_i - \hat{y_i})^2
\end{align}

---

#  Gradient
The gradient of the network is a vector containing the partial derivative of the error with respect to every weight in the network:

\begin{align}
\large
\nabla E = (\frac{\partial E}{\partial W_{i}}, \frac{\partial E}{\partial W_{i+1}}, ... , \frac{\partial E}{\partial W_{n}})
\end{align}

Since the error isn't expressed directly as a function of the weights, we must take a look at how the network is feeding forward and how we can work our way backwards. 

## Finding the partial derivatives
There's two ways you can approach finding $\frac{\partial E}{\partial W^n_{ij}}$

1. Express the error in terms of the given weight (partial equation), then differentiate
2. Use chain rule


For example, using approach 2, if you wanted to to find $\frac{\partial E}{\partial W^2_{11}}$ you work you're way back through the feed forward equations and differentiate each function with respect to the inputs of those functions repeatedly:

\begin{align}
\large
\frac{\partial E}{\partial W^1_{11}} = \frac{\partial E}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial y_i} \frac{\partial y_i}{\partial W^1_{11}}
\end{align}

```python
#### Imports ####
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap 
```

## Tools

```python
colors_list = ['red', 'cyan', 'magenta', 'green', 'black', 'blue']
colors = ListedColormap(colors_list)

def MSE(y, y_hat):
    return np.mean((y-y_hat)**2)

def generate_area_map(features, points_per_int = 10, alpha = 0.2):
    ###  Area Map Set  ###
    xstart = int((features[:, 0].min() - 1) * points_per_int) # min start point of the x data
    xrang = int((features[:, 0].max() + 1) * points_per_int - xstart) # range of the data on x

    ystart = int((features[:, 1].min()  - 1) * points_per_int) # min start poing of the y data
    yrang = int((features[:, 1].max() + 1) * points_per_int - ystart) # range of teh data on y

    # Creates an array with all the coordinates of area map set
    area_map_set = np.array([[x + xstart, y + ystart] for x in range(xrang) for y in range(yrang)])
    area_map_set = area_map_set / points_per_int
    
    return area_map_set

def area_map_plot(network, area_map_set, features, targets, path = '', alpha = 0.1):
    # gets the prediction the model made for the area map set
    pred = network.predict(area_map_set)
    pred = [pred[i,:].argmax() for i in range(int(pred.shape[0]))]

    # draws the current area map and the test set overtop it, saves the scatter
    plt.scatter(features[:, 0], features[:, 1], c = targets, cmap =colors)
    plt.scatter(area_map_set[:, 0], area_map_set[:, 1], c = pred, alpha = alpha, cmap = colors)
    
    if path == '':
        plt.show()
    else:
        plt.savefig(path)
        plt.close()
```

## Multi-Layer Perceptron Network (1 Hidden Layer)

```python
#### Neural Network Class ####
class MLP_Example:
    ##### Constructor ####
    def __init__(self, n_input_nodes, n_hidden_nodes, n_output_nodes, lr):
        ## Network ##
        self.n_input_nodes = n_input_nodes
        self.n_hidden_nodes = n_hidden_nodes
        self.n_output_nodes = n_output_nodes
        
        ## Weights and Biases##
        # Weights and biases initialized with a uniform random distribution
        self.weights_i_h = np.random.uniform(-1.0, 1.0, (self.n_input_nodes, self.n_hidden_nodes))
        self.weights_h_o = np.random.uniform(-1.0, 1.0, (self.n_hidden_nodes, self.n_output_nodes))
        self.bias_i_h = np.random.uniform(-1.0, 1.0, (1, self.n_hidden_nodes))
        self.bias_h_o = np.random.uniform(-1.0, 1.0, (1, self.n_output_nodes))
        
        ## Learning Rate ##
        self.lr = lr
        
        ## Activation Functions ##
        # Linear Activation
        self.relu = lambda x: x
        self.d_relu = lambda x: np.ones(x.shape)
            
        # Sigmoid Activation
        self.sigmoid = lambda x: 1 / (1 + np.exp(-x))
        self.d_sigmoid = lambda x: x * (1 - x)  # assuming sigmoid has already applied to x
    
    #### Feed Forward ####
    def feed_forward(self, X):
        # Hidden Layer
        hidden_inputs = np.dot(X, self.weights_i_h) + self.bias_i_h
        hidden_outputs = self.sigmoid(hidden_inputs)
                
        # Output Layer
        y_hat_raw = np.dot(hidden_outputs, self.weights_h_o) + self.bias_h_o
        y_hat = self.sigmoid(y_hat_raw)
        
        return hidden_outputs, y_hat
    
    #### Backpropagation ####
    def backpropagation(self, X, y, y_hat, hidden_outputs):
        # Output Layer
        d1 = y - y_hat
        d2 = self.d_sigmoid(y_hat)
        output_error = d1 * d2
        
        # Hidden Layer
        d3 = self.weights_h_o
        d4 = self.d_sigmoid(hidden_outputs)
        hidden_error = np.dot(output_error, d3.T) * d4
        
        # Gradients
        output_grad = hidden_outputs.T * output_error
        hidden_grad = X.T * hidden_error
        
        return hidden_grad, hidden_error, output_grad, output_error
    
    #### Training ####
    def train(self, features, targets):
        # Batch Size for weight update step
        batch_size = features.shape[0]
        
        # Delta Weights Variables
        delta_weights_i_h = np.zeros(self.weights_i_h.shape)
        delta_weights_h_o = np.zeros(self.weights_h_o.shape)
        delta_bias_i_h = np.zeros(self.bias_i_h.shape)
        delta_bias_h_o = np.zeros(self.bias_h_o.shape)
        
        # For every data point, forward pass, backpropogation, store weights change
        for X, y in zip(features, targets):
            # Forward pass
            X = X.reshape(1, X.shape[0])
            hidden_outputs, y_hat = self.feed_forward(X)
            
            # Back propogation
            hidden_grad, bias_h_grad, output_grad, bias_o_grad = self.backpropagation(X, y, y_hat, hidden_outputs)
            
            # Weights
            delta_weights_i_h += hidden_grad
            delta_weights_h_o += output_grad
            # Bias
            delta_bias_i_h += bias_h_grad
            delta_bias_h_o += bias_o_grad
        
        
        # Update Weights and biases
        self.weights_i_h += (self.lr * delta_weights_i_h) / batch_size
        self.weights_h_o += (self.lr * delta_weights_h_o) / batch_size 
        self.bias_i_h += (self.lr * delta_bias_i_h) / batch_size
        self.bias_h_o += (self.lr * delta_bias_h_o) / batch_size
    
    #### Testing Methods ####
    def predict(self, X):
        # Gives prediction
        return self.feed_forward(X)[1]
    
    def test(self, features, targets):
        predictions = self.predict(features)

        n_correct = 0
        for i in range(len(predictions)):
            prediction = np.argmax(predictions[i])
            correct = np.argmax(targets[i])

            if prediction == correct:
                n_correct += 1

        return n_correct / len(targets)
```

## Network and Parameters

```python
## Network Parameters ##
n_input_nodes = n_features
n_output_nodes = n_classes
n_hidden_nodes = 128
lr = 0.2

## Neural Network ##
network = MLP_Example(n_input_nodes, n_hidden_nodes, n_output_nodes, lr)
```

### Control Test

```python
area_map_set = generate_area_map(x_test, points_per_int = 7)
area_map_plot(network, area_map_set, x_test, y_test_raw, alpha = 0.15)
    
print("Accuracy on Train Set:", network.test(x_train, y_train))
print("Accuracy on Test Set:", network.test(x_test, y_test))
```

## Train

```python
## Training Params ##
n_epochs = 1000
batch_size = 64 # set to 1 for SGD 

## Tracking loss and accuracy ##
losses = {'train':[], 'validation':[]}
accuracy = {'train':[], 'validation':[]}

## Training Algorithim ##
for epoch in range(n_epochs):
    batch_indices = np.random.choice(list(range(x_train.shape[0])), size = batch_size)
    features = np.array([x_train[i] for i in batch_indices])
    targets = np.array([y_train[i] for i in batch_indices])

    network.train(features, targets)

    # Printing out the training progress
    train_loss = MSE(network.predict(x_train).T, y_train_raw)
    val_loss = MSE(network.predict(x_test).T, y_test_raw)
    train_acc = network.test(x_train, y_train)
    val_acc = network.test(x_train, y_train)
    sys.stdout.write("\rProgress: {:2.1f}%".format(100 * epoch/float(n_epochs)) \
                     + " ... Training loss: " + str(train_loss)[:5] \
                     + " ... Validation loss: " + str(val_loss)[:5] \
                     + " ... Training Acc: " + str(train_acc)[:5] \
                     + " ... Validation Acc: " + str(val_acc)[:5])
    sys.stdout.flush()

    losses['train'].append(train_loss)
    losses['validation'].append(val_loss)
    accuracy['train'].append(train_acc)
    accuracy['validation'].append(val_acc)
```

### Loss Graph

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = False, sharex = False, figsize=(16, 6))

_ = ax1.plot(losses['train'], label='Training loss')
_ = ax1.plot(losses['validation'], label='Validation loss')

_ = ax2.plot(accuracy['train'], label='Training acc')
_ = ax2.plot(accuracy['validation'], label='Validation acc')

_ = fig.legend()
```
